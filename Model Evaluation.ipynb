Metrics

MSE (Mean Squared Error)

```python
def mse(y,x,w,b):
  sum = 0
  for i in range(0,len(x)):
    sum += (y[i] - (w*x[i]+b))**2
  return sum / len(y)
```

RMSE (Root Mean Squared Error)

```python
def rmse(y,x,w,b):
  sum = 0
  for i in range(0,len(x)):
    sum += (y[i] - (w*x[i]+b))**2
  return (sum / len(y))**(1/2)
```

MAE (Mean Absolute Error)

```python
def mae(y,x,w,b):
  sum = 0
  for i in range(0,len(x)):
    sum += abs(y[i] - (w*x[i]+b))
  return sum / len(y)
```
