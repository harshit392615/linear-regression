some sample data

```python
# Example feature matrix (10 samples, 3 features)
x = np.array([
    [1, 2, 3],
    [2, 3, 4],
    [3, 4, 5],
    [4, 5, 6],
    [5, 6, 7],
    [6, 7, 8],
    [7, 8, 9],
    [8, 9, 10],
    [9, 10, 11],
    [10, 11, 12]
], dtype=float)

# Target variable
y = np.array([14, 20, 26, 32, 38, 44, 50, 56, 62, 68], dtype=float)
```

adding ones column in x matrix for bias and adding bias element in bias matrix

```python
ones = np.ones((x.shape[0],1))
x = np.column_stack((ones , x))
w = np.zeros((x.shape[1],1))
y = y.reshape(-1,1)
```

gradient function


```python
def gradient(x,y,w):
  n = x.shape[0]
  dw = 0
  error = x @ w - y
  dw = (2/n)*(x.T @ error) # at this step we calculated both bias and weights
  return dw
```

```python
epochs = 10000
alpha = 0.01

def multi_linear_regression(x,y,w,epochs,alpha):
  for i in range(epochs):
    dw = gradient(x,y,w)

    w = w - alpha*dw # at thia step we will update both bias and weights

  b = w[0]
  w = w[1:]
  return b , w


```

```python
print(multi_linear_regression(x,y,w,epochs,alpha))
```
